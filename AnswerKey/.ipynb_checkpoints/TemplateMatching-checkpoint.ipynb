```python
def fcn_math_operation(math_str, values):
    if math_str == 'plus':
        output = values[0] + values[1]
    elif math_str == 'minus':
        output = values[0] - values[1]
    elif math_str == 'times':
        output = values[0] * values[1]
    else:
        eps = 0.001
        output = values[0] / max(values[1], eps)
    
    return output
```
